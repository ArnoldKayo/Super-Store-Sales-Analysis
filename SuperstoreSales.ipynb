```python
#To get top 5 ship modes by profit
# Group the data by Ship Mode and calculate the total profit for each mode
ship_mode_profit = superstore_sales.groupby("Ship Mode")["Profit"].sum().reset_index()

# Sort the ship modes by profit in descending order and get the top 5
top_5_ship_modes = ship_mode_profit.sort_values(by="Profit", ascending=False).head()

# Print the top 5 ship modes by profit
print(top_5_ship_modes)
```

```python
#To get top 5 categories by sales
# Group the data by Category and calculate the total sales for each mode
categories_sales = superstore_sales.groupby('Category')['Sales'].sum().reset_index()

# Sort the Category by sales in descending order and get the top 5
top_5_categories = categories_sales.sort_values(by='Sales', ascending=False).head()

# Print the top 5 categories by sales
print(top_5_categories)
```

```python
#To get segment by profit
# Group the data by segment and calculate the total profit for each mode
segment_profit = superstore_sales.groupby('Segment')['Profit'].sum().reset_index()

# Sort the segment by profit in descending order
total_profit_by_segment = segment_profit.sort_values(by='Profit', ascending=False)

# Print the total_profit_by_segment
print(total_profit_by_segment)
```
